# **Criando modelo para prever tópicos de planos de ensino**

```python
#Lendo o arquivo e importando as bibliotecas do python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB

dados = pd.read_parquet("/content/teaching-plan-text-classifier/data/df-teaching-plan-labeled.parquet", engine="pyarrow")
dados.sample(11)
```

```python
#Separando os dados em treinamento (60%) e teste (40%)

X_treino, X_teste, y_treino, y_teste = train_test_split(dados['Texto'], dados['Tópico'], test_size=0.4, random_state=42)
vectorizer = TfidfVectorizer()
X_treino2 = vectorizer.fit_transform(X_treino)
X_teste2 = vectorizer.transform(X_teste)
```

# **Usando o LinearSVC para criar o modelo**

```python
#Criando o modelo para o realizar o treinamento

model = LinearSVC()
model.fit(X_treino2, y_treino)
```

```python
#Fazendo o modelo prever os tópicos do texto

def previsao(texto):
    texto_tabela = vectorizer.transform([texto])
    return model.predict(texto_tabela)[0]
texto_entrada = input("Digite o texto:" " ")
previsao_topico = previsao(texto_entrada)
print(f'O tópico do texto é: {previsao_topico}')
```

```python
#Verificação da taxa de acertos do modelo

from sklearn.metrics import accuracy_score
y_previsao = model.predict(X_teste2)
acuracia = accuracy_score(y_teste, y_previsao)
print("Taxa de acerto: %.2f%%" % (acuracia * 100))
```

# **Usando o MultinomialNB para criar o modelo**

```python
#Criando o modelo para o realizar o treinamento

modelo2 = MultinomialNB()
modelo2.fit(X_treino2, y_treino)
```

```python
#Fazendo o modelo prever os tópicos do texto

def previsao(texto):
    texto_tabela = vectorizer.transform([texto])
    return modelo2.predict(texto_tabela)[0]
texto_entrada = input("Digite o texto:" " ")
previsao_topico = previsao(texto_entrada)
print(f'O tópico do texto é: {previsao_topico}')
```

```python
#Verificação da taxa de acertos do modelo MultinomialNB

from sklearn.metrics import accuracy_score
y_previsao = modelo2.predict(X_teste2)
acuracia = accuracy_score(y_teste, y_previsao)
print("Taxa de acerto: %.2f%%" % (acuracia * 100))
```
